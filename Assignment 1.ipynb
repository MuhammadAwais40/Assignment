## 2- create a numpy array of 10 integers. Find mean, median and standard deviation of the array....

```python
import numpy as np

# Create a NumPy array of 10 integers
array = np.array([12, 45, 67, 23, 89, 34, 56, 78, 90, 21])

# Calculate mean, median, and standard deviation
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)

# Display the results
print(f"Array: {array}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

## 3- write a program that input user's marks and display their grade....

```python
def calculate_grade(marks):
    """Function to determine grade based on marks."""
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    else:
        return "F"

while True:
    try:
        # Input marks from user
        marks = int(input("Enter marks (0-100) or type '-1' to exit: "))
        
        # Check for termination condition
        if marks == -1:
            print("Exiting the program. Goodbye!")
            break
        
        # Validate marks
        if 0 <= marks <= 100:
            grade = calculate_grade(marks)
            print(f"Your grade is: {grade}")
        else:
            print("Invalid marks! Please enter a value between 0 and 100.")
    except ValueError:
        print("Invalid input! Please enter a numeric value.")
```

## 4- create a numpy array by using np.linespace() functions....

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array of 100 points between 0 and 2
x = np.linspace(0, 2 * np.pi, 100)

# Compute cosine and sine values
cos_values = np.cos(x)
sin_values = np.sin(x)

# Plot the cosine and sine curves
plt.figure(figsize=(8, 5))
plt.plot(x, cos_values, label='Cosine Curve', color='blue', linestyle='--')
plt.plot(x, sin_values, label='Sine Curve', color='green', linestyle='-')

# Add labels, legend, and title
plt.title('Comparison of Cosine and Sine Curves')
plt.xlabel('x (radians)')
plt.ylabel('Function Value')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## 5- create a mini project to analyze and visualize the dataset....

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.DataFrame({
    "Name": ["Ali", "Ahmad", "Hassan", "Babar", "Amir", "Bilal", "Danish", "Fahad", "Irfan", "Raza"],
    "Math": [78, 85, 62, 90, 88, 76, 65, 92, 81, 87],
    "Science": [85, 80, 72, 95, 90, 70, 60, 93, 82, 86],
    "English": [88, 75, 65, 92, 89, 78, 70, 94, 83, 84]
})

# Display the first few rows of the dataset
print("Dataset Preview:")
print(data)

# Add a column for average percentage
data["Average"] = data[["Math", "Science", "English"]].mean(axis=1)

# Plot histograms for percentage distributions
plt.figure(figsize=(12, 6))
plt.hist(data["Average"], bins=5, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Histogram of Students' Average Percentages")
plt.xlabel("Average Percentage")
plt.ylabel("Frequency")
plt.grid(alpha=0.3)
plt.show()

# Scatter plot of Math vs. Science scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Math", y="Science", data=data, hue="Average", size="Average", sizes=(50, 200), palette="coolwarm")
plt.title("Scatter Plot of Math vs Science Scores")
plt.xlabel("Math Percentage")
plt.ylabel("Science Percentage")
plt.legend(title="Average %")
plt.grid(alpha=0.3)
plt.show()

# Bar plot of average percentages by students
plt.figure(figsize=(10, 6))
sns.barplot(x="Name", y="Average", data=data, palette="viridis")
plt.title("Bar Plot of Average Percentages by Student")
plt.xlabel("Student Name")
plt.ylabel("Average Percentage")
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation = data[["Math", "Science", "English", "Average"]].corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

```
